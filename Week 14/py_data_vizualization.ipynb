```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import numpy.random as npr
#figures and subplots
fig = plt.figure(figsize = (2,4), dpi = 100)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
plt.plot(randn(10)*10)

ax2 = fig.add_subplot(2,2,2)
plt.plot(np.random.random(10)*10)

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randint(1,100, size =10)
         
ax4 = fig.add_subplot(2,2,4)
plt.plot(randn(10)*50)
```

```python
# line color and style
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
plt.plot(npr.randn(10)*10)

ax2 = fig.add_subplot(2,2,2)
plt.plot(npr.random(10)*10)

ax3 = fig.add_subplot(2,2,3)
plt.plot(npr.random(1,100), 'k.')
         
ax4 = fig.add_subplot(2,2,4)
plt.plot(npr.randn(10)*50, 'k--')
```

```python
fig = plt.figure()
dataset = npr.randn(50).cumsum()
```

```python
ax1.plot(dataset, 'k--')

ax2.hist(dataset, bins = 20, color = 'k', alpha = 0.25)

ax3.scatter(np.arange(len(dataset)), dataset.cumsum())

ax4.plot(dataset, 'ko')
plt.show()
```

```python
data = npr.randn(30).cumsum()
plt.plot(data, 'k--', drawstyle = 'steps-post')
plt.plot(data, 'k--')
```

```python
#using labels
plt.plot(data, 'k--', drawstyle = 'steps-post', label = 'steps-post')
plt.plot(data, 'k--', label = 'dashed-line')
plt.legend(loc='best')
```

```python
#plot multiple columns
data = {
    'col1': np.random.randint(1,21, size =20),
    'col2': np.random.randint(20,41, size =20),
    'col3': npr.randn(20),
    'col4': npr.randn(20).cumsum()
}
df = pd.DataFrame(data)
df
```

```python
plt.plot(df['col1'], 'k.', label = 'col1')
plt.plot(df['col2'], 'ko--', label = 'col2')
plt.legend(loc='best')
```

```python
#ticks, labels, and legends
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(npr.randn(1000).cumsum())
```

```python
ax.set_xticks(list(range(0,1001,250)))

#labels
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 30, fontsize = 'small')

#title
ax.set_title('My first matplotlib plot with ticks')
ax.set_xlabel('Stages')
ax.set_ylabel('Values')
fig
```
